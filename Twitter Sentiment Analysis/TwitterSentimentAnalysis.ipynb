```python
def airlineChange(df):
    for i in range(len(df)):
        if(df[i]=='United'):
            df[i]=1
        if(df[i]=='US Airways'):
            df[i]=2
        if(df[i]=='American'):
            df[i]=3
        if(df[i]=='Southwest'):
            df[i]=4
        if(df[i]=='Delta'):
            df[i]=5
        if(df[i]=='Virgin America'):
            df[i]=6
```
