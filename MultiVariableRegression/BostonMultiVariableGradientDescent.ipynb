```python
def step_gradient(x_train,y_train,learning_rate, m , c):
    m_slope = np.zeros(13)
    c_slope = 0
    M = len(x_train)
    for i in range(M):
        for j in range(13):
            m_slope[j] += (-2/M)* (y_train[i] - m[j] * x_train[i][j] - c)*x_train[i][j]
            c_slope += (-2/M)* (y_train[i] - m[j] * x_train[i][j] - c)
    new_m = m - learning_rate * m_slope
    new_c = c - learning_rate * c_slope
    print(cost(x_train,y_train,new_m,new_c))
    return new_m, new_c
```

```python
def cost(x_train,y_train,new_m,new_c):
    M=len(x_train)
    total_cost=0
    for i in range(len(x_train)):
        total_cost=total_cost+(y_train[i]-(new_m*x_train[i]).sum()-new_c)**2
    return total_cost/M
        
```

```python
def gd(x_train,y_train, learning_rate, num_iterations):
    m = np.zeros(13)       
    c = 0       
    for i in range(num_iterations):
        m, c = step_gradient(x_train,y_train, learning_rate, m , c)
        #print(i, " Cost: ", cost(points, m, c))
    return m, c
```

```python
def run():
    learning_rate = 0.01
    num_iterations = 10
    m, c = gd(x_train,y_train, learning_rate, num_iterations)
    print("Final m :", m)
    print("Final c :", c)
    return m,c
```
